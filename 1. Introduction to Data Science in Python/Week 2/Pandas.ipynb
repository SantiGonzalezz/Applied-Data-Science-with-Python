## Manipulating DataFrame

---

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/presidents.csv')
df.head(1)
```

```python
df['Name'] = df['President']
df['Name'] = df['Name'].replace('[ ].*', '', regex=True)
df.head(1)
```

<a id='c-del'></a>

[Índice](#index)

### `del`

```python
del df['Name']
```

<a id='c-apply'></a>

[Índice](#index)

### `apply`

```python
def splitname(row):
    # row is a Series object indexed by column values
    row['First'] = row['President'].split()[0]
    row['Last'] = row['President'].split()[-1]
    
    return row

df = df.apply(splitname, axis='columns')
df.head(1)
```

```python
del df['First']
del df['Last']
```

<a id='c-extract'></a>

[Índice](#index)

### `extract`

```python
# https://stackoverflow.com/questions/10804732/difference-between-and#:~:text=The%20difference%20between%20%3F%3A%20and,%22abc%22%2C%20while%20a(%3F%3D

pattern = '(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)'

names = df['President'].str.extract(pattern)
names.head()
```

```python
df['First'] = names['First']
df['Last'] = names['Last']

df.head(1)
```

```python
pattern = '([\w]{3} [\d]{1,2}, [\d]{4})'
dates = df['Born'].str.extract(pattern)
dates.head(2)
```

```python
df['Born'] = df['Born'].str.extract(pattern)
df.head(1)
```

<a id='c-to-datetime'></a>

[Índice](#index)

### `to_datetime`

```python
# https://dataindependent.com/pandas/pandas-to-datetime-string-to-date-pd-to_datetime/

df['Born'] = pd.to_datetime(df['Born'], format='%b %d, %Y')
df.head(1)
```
